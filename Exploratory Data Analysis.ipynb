```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\varsh\Downloads\GOOG.csv")
df
```

```python
print(df.info())  
print(df.describe())  
print(df.head())  
```

```python
df1 = df['open'].rolling(10).mean()
df1
```

```python
df2 = df1.iloc[-1]
df2
```

```python
df1.plot()
```

```python
numeric_data = df[['open', 'close', 'high', 'low', 'volume']]

# Step 3: Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Step 4: Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Stock Metrics')
plt.show()
```

```python
df['DPR'] = ((df['close'] - df['open']) / df['open']) * 100
df['DPR']
```

```python
df['DPR'].plot()
```
